### Collect Data

```python
competitions_df = sb.competitions()
competitions_df = competitions_df[competitions_df['competition_gender'] == 'male']
competitions_df.head()
```

```python
competitions_df = competitions_df[['country_name', 'competition_name', 'season_name']]
competitions_df.tail()
```

```python
competitions_df.competition_name.value_counts()
```

```python
la_liga_competitions_df = competitions_df[competitions_df['competition_name'] == 'La Liga'].reset_index(drop=True)
champions_league_competitions_df = competitions_df[competitions_df['competition_name'] == 'Champions League'].reset_index(drop=True)

la_liga_competitions_df.shape, champions_league_competitions_df.shape
```

```python
la_liga_competitions_df = la_liga_competitions_df[la_liga_competitions_df['season_name'].isin(champions_league_competitions_df['season_name'])]
champions_league_competitions_df = champions_league_competitions_df[champions_league_competitions_df['season_name'].isin(la_liga_competitions_df['season_name'])]

la_liga_competitions_df.shape, champions_league_competitions_df.shape
```

```python
merged_competitions_df = pd.concat([la_liga_competitions_df, champions_league_competitions_df], ignore_index=True).sample(frac=1).reset_index(drop=True)
merged_competitions_df.head()
```

```python
selected_fields = ['id', 'index', 'period', 'type', 'possession_team', 'play_pattern', 'position', 
                   'location', 'match_id', 'shot_end_location', 'shot_technique', 'shot_outcome']
```

```python
def get_shots_df(request_df: pd.DataFrame, fields: list, drop=False, shuffle=False):
    result_df = pd.DataFrame()
    for _, row in request_df.iterrows():
        try:
            cur_events_df = sb.competition_events(country=row['country_name'],
                                                  division=row['competition_name'],
                                                  season=row['season_name'],
                                                  split = True)['shots']

            cur_events_df = cur_events_df[fields] if not drop else cur_events_df.drop(fields, axis=1)
            cur_events_df['competition_name'] = row['competition_name']
            cur_events_df['season_name'] = row['season_name']

            result_df = pd.concat([result_df, cur_events_df],  ignore_index=True)
        except:
            print('COULDNT PROCESS ROW')
            print(row, 2*'\n')

        
    if shuffle: result_df = result_df.sample(frac=1).reset_index(drop=True)
    
    return result_df
```

```python
shots_df = get_shots_df(merged_competitions_df, 
                                  ['id', 'location', 'match_id', 'shot_end_location', 'shot_outcome'])
```
